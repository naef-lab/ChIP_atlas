```python
import pandas as pd
import numpy as np
import pyBigWig as bw
```

```python
genome = 'mm10'
chip_experiment = f'../resources/experimentList_v3_{genome}_TFs_only_QC_filtered.tab'
chip = pd.read_csv(chip_experiment, sep='\t',index_col=0)

# pick 10 random experiments
chr_size = pd.DataFrame()
for id in chip.sample(1000).index:
    f = bw.open(f'../resources/tracks/{genome}/{id}.bw')
    chr_size = pd.concat([chr_size,pd.DataFrame.from_dict(f.chroms(),orient='index',columns=[id])],axis=1)

N = {}
for chr in chr_size.index:
    vals = chr_size.loc[chr].values
    vals = vals[~np.isnan(vals)].astype(int)
    N[chr] = len(np.unique(vals))
    
    if N[chr] > 1:
        print(chr_size.loc[chr].unique())
```
