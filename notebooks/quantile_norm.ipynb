```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quantile_normalize(a, ref):
    # create the rank array for the array to be normalized, namele the array a
    a_ranks = a.argsort(axis=None).argsort(axis=None)
    # create the normalized array
    ref_s = np.sort(ref, axis=None)
    a_norm=ref_s[a_ranks]
    
    return a_norm
```

```python
n = 10000

a = np.random.exponential(1, n)

def gaussian(x,mu,sigma):
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*((x-mu)/sigma)**2)


ref = np.random.normal(0,1,n)
a_norm = quantile_normalize(a, ref)

fig,axes = plt.subplots(1,3,figsize=(10,5))
axes[0].hist(a, bins=20, alpha=0.5, label='a')
axes[1].hist(ref, bins=20, alpha=0.5, label='a')
axes[2].hist(a_norm, bins=20, alpha=0.5, label='a_norm')
```
